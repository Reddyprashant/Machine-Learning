```python
from __future__ import print_function
import urllib3 #  pip install -i https://pypi.anaconda.org/pypi/simple urllib3
import unicodedata
from bs4 import BeautifulSoup, SoupStrainer, Comment
import warnings
import time
warnings.filterwarnings('ignore')
```

```python
#URL used
url='https://in.bookmyshow.com/'
```

```python
#Checking response status of the above URL whether URL is working or not
http = urllib3.PoolManager()
response = http.request('GET',url)
response.status
```

```python
soup=BeautifulSoup(response.data, "lxml")
```

```python
#Finding the Data from html from the tag (<a)
soup.find_all('a')
```

```python
#Getting the text present between the tag 'a'(<a) from the html 
text = ' '.join(map(lambda p: p.text, soup.find_all('a')))
text
```

```python
#Finding the tags which has href
soup.find_all('a', href=True)
```

```python
for a1 in soup.find_all('a', href=True):
 print(a1['href'])
```

```python
#Finding the External Links which are present in website
for a1 in soup.find_all('a',href=True):
    if('http' in a1['href']):
     print(a1['href'])
```

```python
dictionary=dict();
for a2 in soup.find_all('a',href=True):
    if('http' in a2['href']):
        dictionary[a2['href']]=a2.text
print (dictionary)
        
```

```python
#Exporting external links, URL response and time. to a .csv file
http = urllib3.PoolManager()
t=''
import csv
with open('BookMyShow.csv', 'w') as csvfile:
    fieldnames = ['Link', 'Data','Link Status', 'Time']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for a4 in dictionary.keys(): 
        try:
            response = http.request('Get',a4);
            if(response.status==200):
                print(a4)
                t='Link Is Working Fine'
                writer.writerow({'Link': a4, 'Data':dictionary[a4], 'Link Status': t , 'Time':time.strftime('%c')});
        
            else:
                print(a4)
                t='Link is Not Working'
                writer.writerow({'Link': a4, 'Data':dictionary[a4], 'Link Status': t, 'Time':time.strftime('%c')});
            
        except:
                print(a4)
                t=print('Link is Not Working')
                writer.writerow({'Link': a4, 'Data':dictionary[a4], 'Link Status': t, 'Time':time.strftime('%c')});
            
          
```
